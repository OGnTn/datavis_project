## Processing

```python
#Drop all columns except for name and location
df_universities = df_universities[['University_Name', 'Location']]
```

```python
#sepate location into city and country
df_universities['City'] = df_universities['Location'].str.split(',').str[0]
df_universities['Country'] = df_universities['Location'].str.split(',').str[1]
df_universities = df_universities[['University_Name', 'City', 'Country']]
df_universities.head(3)
```

```python
#drop duplicates
print(df_universities.shape)
df_universities.drop_duplicates(inplace=True)
print(df_universities.shape)
```

```python
#Drop all columns except for university name, country, city, lang., and tuition/semester
df_masters = df_masters[['UNIVERSITY NAME', 'COUNTRY', 'CITY', 'LANG.']]
df_masters.head(3)
```

```python
#drop duplicates
print(df_masters.shape)
df_masters.drop_duplicates(inplace=True) 
print(df_masters.shape)
```

```python
#renaming columns
df_masters.rename(columns={'UNIVERSITY NAME': 'University_Name', 'COUNTRY': 'Country', 'CITY': 'City', 'LANG.': 'Language'}, inplace=True)
df_masters.head(3)
#add language column to universities df

print(df_masters.columns)

#switch column positions in universities df
df_universities = df_universities[['University_Name', 'Country', 'City']]
print(df_universities.columns)
df_universities.head(3)
df_merged = pd.concat([df_universities, df_masters])
df_merged.head(3)

df_merged.shape
```

```python
df_cost = df_cost[['country', 'city', 'x1', 'x4', 'x5','x25','x26','x16', 'x17','x18','x19','x20','x21','x23','x27','x29','x31','x36','x38','x41','x48','x49']]
df_cost.rename(columns={'x1': 'meal_inexpensive_restaurant', 'x23': 'water', 'x27': 'cigarettes', 'x29': 'monthly_pass', 'x31': 'taxi', 'x36': 'basic', 'x38': 'internet','x41': 'cinema', 'x41':'cinema', 'x48': 'appartment_centre', 'x49': 'appartment_outside_centre' }, inplace=True)
#add column beer_restaurant which is the mean of the x4 and x5 columns and drop them
df_cost['beer_restaurant'] = df_cost[['x4', 'x5']].mean(axis=1)
df_cost.drop(['x4', 'x5'], axis=1, inplace=True)
#add column beer_market which is the mean of the x25 and x26 columns and drop them
df_cost['beer_market'] = df_cost[['x25', 'x26']].mean(axis=1)
df_cost.drop(['x25', 'x26'], axis=1, inplace=True)
#add column fruits(1kg) which is the mean of the x16, x17, x18, x19 columns and drop them
df_cost['fruits'] = df_cost[['x16', 'x17', 'x18', 'x19']].mean(axis=1)
df_cost.drop(['x16', 'x17', 'x18', 'x19'], axis=1, inplace=True)
#add column vegetables(1kg) which is the mean of the x20, x21 columns and drop them
df_cost['vegetables'] = df_cost[['x20', 'x21']].mean(axis=1)
df_cost.drop(['x20', 'x21'], axis=1, inplace=True)

```
